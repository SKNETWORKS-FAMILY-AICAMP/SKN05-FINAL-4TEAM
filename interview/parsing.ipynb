```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
```

```python
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
url = "https://www.jobkorea.co.kr/Recruit/GI_Read/46006552?Oem_Code=C1&logpath=1&stext=%EB%B0%B1%EC%97%94%EB%93%9C&listno=4"
driver.get(url)
soup = BeautifulSoup(driver.page_source, 'html.parser')
link_tag = soup.find('link', {"media": "only screen and (max-width: 640px)", 'rel': 'alternate'})
driver.get(str(link_tag['href']))
iframe = driver.find_element(By.ID, 'dev_detailifrm')
driver.switch_to.frame(iframe)
soup = BeautifulSoup(driver.page_source, 'html.parser')
print(soup)
# row_embed = soup.select_one("#rowEmbed")
# print(row_embed)
driver.quit()
```
